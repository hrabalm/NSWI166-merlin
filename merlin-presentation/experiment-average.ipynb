```python
import pandas as pd
from models_xgbr import load_and_preprocess_movielens_df
```

```python
train, test = load_and_preprocess_movielens_df()
```

```python
average_ratings = train.groupby("movieId").mean()
average_of_all = train["rating"].mean()
```

```python
average_ratings
```

```python
average_of_all
```

```python
naive_predictions = test.join(average_ratings, on="movieId", how="left", rsuffix="_other").fillna(average_of_all).drop(["genres", "userId_other"], axis=1).rename({"rating_other": "preds"}, axis=1)
naive_predictions
```

```python
naive_predictions.to_csv("average_preds.csv")
```
